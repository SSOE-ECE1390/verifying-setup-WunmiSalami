* Load an (Grayscale) image  (‘Data/Apollo_11_Launch.jpg’)
* View image
* Crop image
* Edit brightness of image
* Rotate image
* Save modified image

```python
import cv2
import matplotlib.pyplot as plt

# Load Apollo image
apollo_image = cv2.imread("Data/Apollo_11_Launch.jpg", cv2.IMREAD_GRAYSCALE)

# Display image
plt.imshow(apollo_image, cmap='gray')

#Get image dimensions
print(apollo_image.shape)

#crop image
cropped_image = apollo_image[100:700, 200:900]

#Edit brightness
lower_brightness = cropped_image - 15

#Rotate image
rotated_image = cv2.rotate(lower_brightness, cv2.ROTATE_90_CLOCKWISE)

#save final Image 
cv2.imwrite("Data/Edited_Apollo_Image.jpg", rotated_image)




```
